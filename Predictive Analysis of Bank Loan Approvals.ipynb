#### Identifying Missing Values:

```python
import matplotlib.pyplot as plt

# Calculate percentage of missing values for each column
app_data_mis_vals = round(data.isnull().sum() / len(data) * 100, 2)

# Plot histogram
plt.figure(figsize=[10, 8])
plt.hist(app_data_mis_vals, bins=20, edgecolor='black')
plt.xlabel('Percentage of Missing Values')
plt.ylabel('Number of Columns')
plt.title('Distribution of Missing Values across Columns')
plt.grid(True)
plt.show()
```

`NOTE :` Handling columns with a high percentage of missing values is crucial for maintaining the quality of your dataset. Here's how you can modify the approach to handle such columns:

Identify Columns with High Missing Values Percentage:

Calculate the percentage of missing values for each column.
Identify columns where the percentage of missing values is greater than 40%.
Drop Columns with High Missing Values:

Remove columns where the percentage of missing values exceeds the threshold (40% in this case).
Dropping these columns ensures that you retain the integrity of the dataset and prevent introducing bias or noise into the analysis.

### Dropping Missing Values

```python
# Dropping Columns with High Missing Values
# List of columns with missing values percentage greater than 40
drop_cols = app_data_mis_vals[app_data_mis_vals > 40].index.tolist()

print('Number of Columns having missing values  greater than 40%:', len(drop_cols))
```

## 4.Data Cleaning

***
`Analysis:` As 99% of the values in `AMT_REQ_CREDIT_BUREAU_HOUR`, `AMT_REQ_CREDIT_BUREAU_DAY`, `AMT_REQ_CREDIT_BUREAU_MON`,
`AMT_REQ_CREDIT_BUREAU_WEEK`,`AMT_REQ_CREDIT_BUREAU_QRT` columns are 0, hence it is safe to impute missing values with 0
For `AMT_REQ_CREDIT_BUREAU_YEAR` column we can impute missing values with the median number of yearsMissing values for columns AMT_REQ_CREDIT_BUREAU_HOUR, AMT_REQ_CREDIT_BUREAU_DAY, AMT_REQ_CREDIT_BUREAU_MON, AMT_REQ_CREDIT_BUREAU_WEEK, and AMT_REQ_CREDIT_BUREAU_QRT are imputed with 0, as 99% of the values in these columns are already 0.
Missing values for the AMT_REQ_CREDIT_BUREAU_YEAR column are imputed with the median value of the column.
***

```python
# Replacing null values with appropriate values for specified columns
replace_values = {
    'AMT_REQ_CREDIT_BUREAU_HOUR': 0,
    'AMT_REQ_CREDIT_BUREAU_DAY': 0,
    'AMT_REQ_CREDIT_BUREAU_WEEK': 0,
    'AMT_REQ_CREDIT_BUREAU_MON': 0,
    'AMT_REQ_CREDIT_BUREAU_QRT': 0,
    'AMT_REQ_CREDIT_BUREAU_YEAR': data['AMT_REQ_CREDIT_BUREAU_YEAR'].median()
}

data.fillna(replace_values, inplace=True)

print('Null Values replaced Successfully')
```

```python
import matplotlib.pyplot as plt

def pie_plot(col_name):
    plt.figure(figsize=[20, 6])

    # Plot for target 1 (Defaulters)
    plt.subplot(1, 2, 1)
    defaulters_palette = sns.color_palette('viridis', len(target_1[col_name].unique()))
    col_pie_1 = target_1[col_name].value_counts(normalize=True).rename_axis('Type').reset_index(name='Count')
    plt.pie(data=col_pie_1, x=col_pie_1.Count, labels=col_pie_1.Type, autopct='%1.1f%%', colors=defaulters_palette)
    plt.title('Pie Chart for Column ' + col_name + ' [For Defaulters]')

    # Plot for target 0 (Non-Defaulters)
    plt.subplot(1, 2, 2)
    non_defaulters_palette = sns.color_palette('mako', len(target_0[col_name].unique()))
    col_pie_0 = target_0[col_name].value_counts(normalize=True).rename_axis('Type').reset_index(name='Count')
    plt.pie(data=col_pie_0, x=col_pie_0.Count, labels=col_pie_0.Type, autopct='%1.1f%%', colors=non_defaulters_palette)
    plt.title('Pie Chart for Column ' + col_name + ' [For Non-Defaulters]')
    plt.show()
```

```python
import matplotlib.pyplot as plt

def barp_plot(col_name):
    plt.figure(figsize=[22, 10])

    # Plot for target 1 (Defaulters)
    plt.subplot(1, 2, 1)
    defaulters_palette = sns.color_palette('rocket', len(target_1[col_name].unique()))
    target_1[col_name].value_counts().plot(kind='barh', color=defaulters_palette)
    plt.title('Bar Chart for Column ' + col_name + ' [For Defaulters]', fontsize=20, fontweight='bold')
    plt.xlabel('Number of Loan Applications', fontsize=16, fontweight='bold')
    plt.ylabel(col_name, fontsize=16, fontweight='bold')

    # Plot for target 0 (Non-Defaulters)
    plt.subplot(1, 2, 2)
    non_defaulters_palette = sns.color_palette('muted', len(target_0[col_name].unique()))
    target_0[col_name].value_counts().plot(kind='barh', color=non_defaulters_palette)
    plt.title('Bar Chart for Column ' + col_name + ' [For Non-Defaulters]', fontsize=20, fontweight='bold')
    plt.xlabel('Number of Loan Applications', fontsize=16, fontweight='bold')
    plt.ylabel(col_name, fontsize=16, fontweight='bold')

    plt.tight_layout()
    plt.show()
```
